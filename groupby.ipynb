```python
def grouper(df,idx,col):
    if 80<= df[col].loc[idx]<=90:
        return "80-90"
    elif 50<=df[col].loc[idx]<=60:
        return "50-60"
    else:
        return "others"
```

```python
g=df.groupby(lambda indx:grouper(df, indx, "temperature"))
g
```

```python
for i,d in g:
    print("i",i)
    print("data",d)
```

```python
def grouper(df,idx,col):
    if 1<=df[col].loc[idx]<=3.9:
        return"Poor"
    elif 4<=df[col].loc[idx]<=7.9:
        return "Average"
    elif 8<= df[col].loc[idx]<=10:
        return"Good"
    else:
        return "others"
```

```python
g=df.groupby(lambda x:grouper(df,x,"imdb_rating"))
g
```
