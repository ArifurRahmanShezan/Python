```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
df_sales=pd.read_excel("D:\\personal\\Data Science\\8_matplotlib_seaborn\\linechart.xlsx")
df_sales
```

```python
plt.plot(df_sales["Quarter"],df_sales["Fridge"])
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.plot(df_sales["Quarter"],df_sales["Fridge"],color="pink",label="Fridge")
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"],color="maroon",label="Dishwasher")
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"],color="yellow",label="Washing Machine")
plt.title("Products Sale")
plt.xlabel("Financials Quarter")
plt.ylabel("Revenue")
plt.legend()
plt.show()
```

```python
total_sales=df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
```

```python
total_sales.index
```

```python
plt.pie(total_sales,labels=total_sales.index,autopct="%1.1f%%",explode=(0.1,0.1,0.1),shadow=True,startangle=140)
plt.show()
```
